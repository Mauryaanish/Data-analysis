# What are the Top 5 most installed Apps in Each popular Categories ??

```python
df.head()
```

```python
df3 =  df.groupby(["Category" , "App"])["Installs"].sum().reset_index()
```

```python
df3 =df3.sort_values("Installs" , ascending = False)
```

```python
df3[df3["Category"] == "GAME"][0:5]
```

```python
df3[df3["Category"] == "PRODUCTIVITY"][0:5]
```

```python
df3[df3["Category"] == "FAMILY"][0:5]
```

```python
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = df3[df3.Category == app]
    df5 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df5 , x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```
